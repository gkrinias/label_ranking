```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
def plot(dimensionality='SFN', noise_type='additive', noise_measurement='alpha'):
  noise_measurements = []

  for i in range(50):
    with open(f'datasets/{dimensionality}/{noise_type}/{noise_measurement}_{i}.pickle', 'rb') as handle:
      data = pickle.load(handle)
    noise_measurements.append(data[noise_measurement])

  noise_measurements = sorted(noise_measurements)

  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_dt.pickle', 'rb') as handle:
    pairwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_dt.pickle', 'rb') as handle:
    labelwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_rf.pickle', 'rb') as handle:
    pairwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_rf.pickle', 'rb') as handle:
    labelwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_hh.pickle', 'rb') as handle:
    pairwise_hh = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_lr.pickle', 'rb') as handle:
    labelwise_lr = pickle.load(handle)

  plt.rcParams['text.usetex'] = True
  plt.figure() # (figsize=(16, 6))
  plt.title(dimensionality + ' ' + noise_type)
  plt.plot(noise_measurements, labelwise_lr)
  plt.plot(noise_measurements, labelwise_dt)
  plt.plot(noise_measurements, labelwise_rf)
  plt.plot(noise_measurements, pairwise_hh)
  plt.plot(noise_measurements, pairwise_dt)
  plt.plot(noise_measurements, pairwise_rf)
  plt.legend([
    'LWLR',
    'LWDT',
    'LWRF',
    'PWHH',
    'PWDT',
    'PWRF',
    ], 
    fontsize=12
  )
  plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
  plt.xlabel(f"$\{noise_measurement}$")
  plt.ylabel('Mean Kendall rank correlation coefficient')
  plt.grid(visible=True, which='major', color='0.8', linestyle='-')
  plt.grid(visible=True, which='minor', color='0.9', linestyle='--')
  plt.minorticks_on()
  plt.show()
```

```python
def plot2(dimensionality='SFN', noise_type='additive'):

  noise_measurements = []

  noise_measurement = 'alpha'

  for i in range(50):
    with open(f'datasets/{dimensionality}/{noise_type}/{noise_measurement}_{i}.pickle', 'rb') as handle:
      data = pickle.load(handle)
    noise_measurements.append(data[noise_measurement])

  noise_measurements = sorted(noise_measurements)

  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_dt.pickle', 'rb') as handle:
    pairwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_dt.pickle', 'rb') as handle:
    labelwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_rf.pickle', 'rb') as handle:
    pairwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_rf.pickle', 'rb') as handle:
    labelwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_hh.pickle', 'rb') as handle:
    pairwise_hh = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_lr.pickle', 'rb') as handle:
    labelwise_lr = pickle.load(handle)

  plt.rcParams['text.usetex'] = True
  
  fig = plt.figure(figsize=(8, 4))
  # fig.suptitle('SFN datasets with Gaussian additive noise', fontsize=12)
  # plt.title(dimensionality + ' ' + noise_type)

  fig.add_subplot(1, 2, 1) # .title.set_text(dimensionality + ' ' + noise_type)
  # plt.title(dimensionality + ' ' + noise_type)
  plt.plot(noise_measurements, labelwise_lr)
  plt.plot(noise_measurements, labelwise_dt)
  plt.plot(noise_measurements, labelwise_rf)
  plt.plot(noise_measurements, pairwise_hh)
  plt.plot(noise_measurements, pairwise_dt)
  plt.plot(noise_measurements, pairwise_rf)
  plt.legend([
    'LWLR',
    'LWDT',
    'LWRF',
    'PWHH',
    'PWDT',
    'PWRF',
    ],
    #title=dimensionality,
    framealpha=0.6,
    loc=3,
    fontsize=10
  )
  plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
  plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
  plt.xlabel(f"$\{noise_measurement}$")
  plt.ylabel("Mean KT coefficient")
  plt.grid(visible=True, which='major', color='0.8', linestyle='-')
  plt.grid(visible=True, which='minor', color='0.9', linestyle='--')
  plt.minorticks_on()

  noise_measurements = []

  noise_measurement = 'beta'

  for i in range(50):
    with open(f'datasets/{dimensionality}/{noise_type}/{noise_measurement}_{i}.pickle', 'rb') as handle:
      data = pickle.load(handle)
    noise_measurements.append(data[noise_measurement])

  noise_measurements = sorted(noise_measurements)

  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_dt.pickle', 'rb') as handle:
    pairwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_dt.pickle', 'rb') as handle:
    labelwise_dt = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_rf.pickle', 'rb') as handle:
    pairwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_rf.pickle', 'rb') as handle:
    labelwise_rf = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_pairwise_hh.pickle', 'rb') as handle:
    pairwise_hh = pickle.load(handle)
  with open(f'results/{dimensionality}/{noise_type}/{noise_measurement}_labelwise_lr.pickle', 'rb') as handle:
    labelwise_lr = pickle.load(handle)

  plt.rcParams['text.usetex'] = True

  fig.add_subplot(1, 2, 2) #.title.set_text(dimensionality + ' ' + noise_type)
  # plt.title(dimensionality + ' ' + noise_type)
  plt.plot(noise_measurements, labelwise_lr)
  plt.plot(noise_measurements, labelwise_dt)
  plt.plot(noise_measurements, labelwise_rf)
  plt.plot(noise_measurements, pairwise_hh)
  plt.plot(noise_measurements, pairwise_dt)
  plt.plot(noise_measurements, pairwise_rf)
  plt.legend([
    'LWLR',
    'LWDT',
    'LWRF',
    'PWHH',
    'PWDT',
    'PWRF',
    ],
    #title=dimensionality,
    framealpha=0.6,
    loc=4,
    fontsize=10
  )
  plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
  plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
  plt.xlabel(f"$\{noise_measurement}$")
  # plt.ylabel('Mean Kendall rank correlation coefficient')
  plt.grid(visible=True, which='major', color='0.8', linestyle='-')
  plt.grid(visible=True, which='minor', color='0.9', linestyle='--')
  plt.minorticks_on()
  
  plt.savefig(f'plots/{dimensionality}_{noise_type}.pdf')
  plt.show()
```
