```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from label_ranking import *
from massart import *
from lr_regressors import *
from LabelRankers import *


d = 10  # data dimension
k = 5  # labels

# Feature vectors distribution
D = multivariate_normal(mean=np.zeros(d), cov=np.identity(d))

# Linear Sorting Function
W_opt = np.random.rand(k, d)

# Train data
X_train = D.rvs(size=5000)

R_train = ground_truth_permutations(X_train, W_opt)
P_train = np.argsort(R_train, axis=1)

# Test data
X_test = D.rvs(size=1000)

R_test = ground_truth_permutations(X_test, W_opt)
P_test = np.argsort(R_test, axis=1)
```

```python
eta_max = 0.1
R_train_noisy = np.array([addNoise(y, eta_max) for y in R_train])
P_train_noisy = np.argsort(R_train_noisy, axis=1)

clf = LabelwiseDecisionTreeLR().fit(X_train, P_train_noisy)
print("LabelwiseDecisionTreeLR:", score(P_test, clf.predict(X_test)))

clf = LabelwiseRandomForestLR().fit(X_train, P_train_noisy)
print("LabelwiseRandomForestLR:", score(P_test, clf.predict(X_test)))

clf = PairwiseDecisionTreeLR().fit(X_train, P_train_noisy)
print("PairwiseDecisionTreeLR:", score(P_test, clf.predict(X_test)))

clf = PairwiseRandomForestLR().fit(X_train, P_train_noisy)
print("PairwiseRandomForestLR:", score(P_test, clf.predict(X_test)))

clf = PairwiseHalfspaceLR(0.005, 0.1, 0.8).fit(X_train, P_train_noisy)
print("PairwiseHalfspaceLR:", score(P_test, clf.predict(X_test)))
```
